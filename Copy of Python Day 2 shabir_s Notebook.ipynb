### While Loop

```python
# Check condition until its false.
# While loop will execute block of code many times and check condition in each iteration.
```

```python
password = 1234
user_entry = int(input("Please enter your password: "))
if password == user_entry:
  print("Login succesful.")

plays = 3

while plays > 0:
  print("Login failed. Try again.")
  user_entry = int(input("Please enter correct password:"))
  if password == user_entry:
    print("Login succesful.")
    break
  plays = plays -1
```

```python
secret_number = 8
tries_left = 3

while tries_left > 0:
    guess = int(input("Guess the secret number (1-10): "))

    if guess > 10:
      print("Please enter a number between 1 to 10")


    if guess == secret_number:
        print("🎉 Correct! You win!")
        break
    elif guess < secret_number:
        print("Too low.")

    else:
        print("Too high.")

    tries_left -= 1 # same as tries_left = tries_left -1
    print(f"Tries left: {tries_left}")

if tries_left == 0:
    print("😢 Game over. The correct number was", secret_number)
```

```python
import random
secret_number = random.randint(1,10) # this is generating a random number between 1 and 10

tries_left = 3

while tries_left > 0:
    guess = int(input("Guess the secret number (1-10): "))

    if guess > 10:
      print("Please enter a number between 1 to 10")

    if guess == secret_number:
        print("🎉 Correct! You win!")
        break
    elif guess < secret_number:
        print("Too low.")

    else:
        print("Too high.")

    tries_left -= 1 # same as tries_left = tries_left -1
    print(f"Tries left: {tries_left}")

if tries_left == 0:
    print("😢 Game over. The correct number was", secret_number)
```
